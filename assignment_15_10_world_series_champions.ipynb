```python
teamslist = []
with open('WorldSeriesWinners.txt') as my_file_obj:
  for each_line in my_file_obj:
    print(each_line)
    teamslist.append(each_line.rstrip())
x = str(input("Please type in a team name"))

Count = 0

for team in teamslist:
  if team == x:
    Count = Count+1
print(x + str(Count+1))
```
